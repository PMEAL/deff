```python
import taichi as ti
import porespy as ps
import matplotlib.pyplot as plt
from deff import compute_effective_diffusivity, solve_diffusion
```

```python
# Tell taichi how to run
ti.init(arch=ti.gpu)
```

```python
# Generate a test image
im = ps.generators.cylinders([400, 400, 200], r=10, porosity=0.7, seed=1)
```

```python
ax = "x"
soln = solve_diffusion(im=im, direction=ax, tol=1e-2)
compute_effective_diffusivity(soln._last_vtr, direction=ax)
```

```python
ax = "y"
soln = solve_diffusion(im=im, direction=ax, tol=1e-2)
compute_effective_diffusivity(soln._last_vtr, direction=ax)
```

```python
ax = "z"
soln = solve_diffusion(im=im, direction=ax, tol=1e-2)
compute_effective_diffusivity(soln._last_vtr, direction=ax)
```
